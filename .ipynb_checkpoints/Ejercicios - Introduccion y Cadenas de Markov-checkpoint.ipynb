4. Supongamos que la persona que tira la moneda lo hace cada vez más rápido (por ejemplo, le lleva tiempo $1/k$ y no $1$), pero la moneda tiene cada vez menos probabilidad de salir cara. Más formalmente, supongamos que:
$$k\to\infty \quad \text{ y } \quad kp \to\lambda$$

Simular la variable $Y= X/k$ con $X\sim Geo(kp)$ y mostrar que para $k$ grande y $kp\to\lambda$, la distribución converge a una exponencial.

### Parte 1
### Definicion de la funcion 

```julia
using Distributions, Plots, StatsBase

function funcTiros(p)
    dist = Bernoulli(p)
    res = 0
    cont = 0
    while res == 0
        cont = cont + 1
        res = rand(dist)
    end
    return cont;
end  
```

### Parte 2

```julia
prob = 0.5
rep = 1000
res = zeros(rep);
for i = 1:1:rep
    res[i] = funcTiros(prob)
    
end
histY = fit(Histogram,res,(0:50))
plot(histY,label="Histograma Y")
```

```julia
t = 0:0.5:50
ecdfY = ecdf(res)

plot(t,ecdfY.(t),label="Distribución empírica")
plot!(t,cdf.(Geometric(prob),t.-1),label="Distribución teórica")
```

### Parte 3

```julia
function funcExitos(p,n)
    num = 0
    cont = 0
    while num < n
        num = num + funcTiros(p)
        cont = cont + 1
    end
    return cont;
end 

prob = 0.2
N = 100
rep = 1000
resB = zeros(rep);
for i = 1:1:rep
    resB[i] = funcExitos(prob,N)
    
end
histY = fit(Histogram,resB,(0:N))
plot(histY,label="Histograma Y")
```

```julia
t = 0:0.5:N
ecdfY = ecdf(resB)

plot(t,ecdfY.(t),label="Distribución empírica")
plot!(t,cdf.(Binomial(N,prob),t),label="Distribución teórica")
```

### Parte 4

```julia
using Distributions

function funcMarkov(P,pi, Tfinal)
    
    N = size(P,1) #Cantidad de estados = cantidad de filas de P
    E=zeros(Tfinal) #Vector para estados
    
    dis_pi = Categorical(vec(pi)) #Estado inicial 
    estado = rand(dis_pi)
    E[1] = estado
    
    for n=2:Tfinal
        
        dis_salto = Categorical(vec(P[estado,:])) #Distribución de salto
        estado = rand(dis_salto)
        E[n] = estado
    end
    
    return E
end


#Utilización de la función definida
Tfinal = 1000
pi_ini = [1 0 0 0]

X=funcMarkov(P,pi_ini, Tfinal)

k=1:maximum(X);
pi_est=zeros(size(k));

for j=1:length(k)
    pi_est[j]=sum(X.==k[j])/Tfinal;
end

print("Resultado del calculo: ")
println(pi)
print("Resultado de la simulación es: ")
println(pi_est)
```

## Ejercicio 3

Modifique el paseo al azar de tiempo discreto para que la probabilidad de ir de $k\to k+1$ sea $p$ y $k\to k-1$ sea $q$ con $p$ y $q$ tales que $p+q\leqslant 1$. Utilizando las simulaciones, identifique cuál es la condición para que exista régimen.

$$P = \begin{pmatrix}
q & p & 0 & 0 & \cdots \\
q & 0 & p & 0 & \cdots \\
0 & q & 0 & p & \cdots \\
\vdots & \vdots & \vdots & \vdots & \ddots 
\end{pmatrix}$$

De las ecuaciones de balance se concluye que:

$$\pi_k p = \pi_{k+1} q \quad \forall k\geqslant 0.$$

Levantando la recurrencia se tiene que:

$$\pi_k = \pi_0 \left(\frac{p}{q}\right)^k.$$

Finalmente, aplicando el mismo procedimiento que en clase y suponiendo que converge:

$$\pi_0 = \frac{q-p}{q} \quad \pi_k = \left(\frac{q-p}{q}\right)\left(\frac{p}{q}\right)^k.$$

```julia
pi_k(k,p,q) = (q-p)/q * (p/q)^k

p=0.35
q=0.36
k=(0:20) #Vector de k

#Si q > p
print("Si q > p, entonces: ")
plot(k,pi_k.(k,0.3,0.5),line=:stem, marker=:circle)
```

```julia
#Si q = p
print("Si q = p, entonces: ")
plot(k,pi_k.(k,0.3,0.3),line=:stem, marker=:circle)
```

```julia
#Si q < p
print("Si q < p, entonces: ")
plot(k,pi_k.(k,0.49,0.1),line=:stem, marker=:circle)
```

Un sistema de transmisión con control de congestión por ventanas funciona del siguiente modo. En cada
transmisión de una ventana (tiempo discreto) transmite $W = 2^n$ paquetes dependiendo del estado. Dichos
paquetes pueden ser recibidos con éxito (no hay congestión) con probabilidad $p$, o perderse alguno con
probabilidad $1 − p$, en cuyo caso se baja la ventana a $W = 1$ y se comienza nuevamente. Si se llega a
$W = 8$ se mantiene en ese nivel mientras no pierda paquetes. Sea $X_k$ el estado actual de transmisión.

 1. Modele $X_k$ como una cadena de Markov de tiempo discreto. Dé el diagrama de estados y transiciones y la matriz $P$ de la cadena.
 2. Discuta en qué casos la cadena es irreducible y aperiódica.
 3. Hallar la distribución invatiante en aquellos casos que exista.
 4. Calcule, en función de la prob. de congestión $p$:
       1. La proporción de tiempo que el sistema logra trabajar al máximo de capacidad.
       2. La cantidad media de paquetes por slot que el sistema logra transmitir (ventana media, $\bar{W}$).

 5. Simule el caso $p=1/2$, muestre que la ventana media es $\bar{W}=2.5$ y compare con la simulación.


### Parte 1

El diagrama de estados y transiciones es: ![Ejemplo](./Imagenes/Ejercicio4_1.jpg)

La matriz P de la cadenta es: $$P=\begin{pmatrix} 1-p & p & 0 & 0 \\ 1-p & 0 & p & 0 \\ 1-p & 0 & 0 & p \\ 1-p & 0 & 0 & p \end{pmatrix}.$$

### Parte 2
En cualquiera de los casos  $p = 1$  o  $p = 0$  la cadena no sera irreducible.

La cadena es aperiodica para cualquier valor de $p$

### Parte 3

Para hallar la distribucion invariante: 

$$\pi: \left\{\begin{array}{cc}\pi = \pi P \\ \pi \mathbf{1} = 1\end{array}\right.$$

$$\pi_1=(1-p)\pi_1+(1-p)\pi_2+(1-p)\pi_3+(1-p)\pi_4\\ 
\pi_2=p\pi_1 \\
\pi_3=p\pi_2 \\
\pi_4=p\pi_3+p\pi_4$$
Donde se obtiene

$$
\pi_2 = p\pi_1 \\
\pi_3 = p^2\pi_1 \\
\pi_4 = \frac{p^3\pi_1}{1-p}
$$ 

Entonces
$$\pi_1 + \pi_2 + \pi_3 + \pi_4 = 1 \quad\rightarrow\quad\pi_1 + p\pi_1+ p^2\pi_1+ \frac{p^3\pi_1}{1-p} = 1\\
\pi_1\left(1+ p+ p^2+ \frac{p^3}{1-p}\right) = 1 \quad \pi_1((1-p)+ p(1-p)+ p^2(1-p)+ p^3) = 1-p \\
$$ 
Finalmente
$$
\pi_1 = 1-p \\
\pi_2 = p(1-p) \\
\pi_3 = p^2(1-p) \\
\pi_4 = p^3
$$

### Parte 4

La proporcion del tiempo que el sistema trabaja a **maxima capaxidad** es equivalente a el tiempo que pasa en el estado más alto **3**.\
En este caso es $\ p^3$.

Para el calculo de $\ \overline{W} \ $ se hace:
$$
2^0(1-p) + 2^1p(1-p) + 2^2p^2(1-p) + 2^3p^3 =\\
(1-p) + 2p(1-p) + 4p^2(1-p) + 8p^3 = \\
1 - p + 2p - 2p^2 + 4p^2 - 4p^3 + 8p^3 =\\ 
1 + p + 2p^2 + 4p^3 = \overline{W}
$$

### Parte 5

```julia
Tfinal = 10000
pi_ini = [1 0 0 0]
p = 0.5

pi_teo = [1-p p*(1-p) p^2*(1-p) p^3]

P_4 = [1-p p 0 0; 1-p 0 p 0; 1-p 0 0 p; 1-p 0 0 p]
X=funcMarkov(P_4,pi_ini, Tfinal)

k=1:maximum(X);
pi_est=zeros(size(k));

for j=1:length(k)
    pi_est[j]=sum(X.==k[j])/Tfinal;
end

W_teo = 1+p+2*p^2+4*p^3
W_est = sum(pi_est.*2.0.^(0:3))

print("Resultado del calculo de invariante: ")
println(pi_teo)
print("Resultado de la simulacion: ")
println(pi_est)
print("Resultadoo del calculo de W teorico: ")
println(W_teo)
print("Resultado de la simulacion da: ")
println(W_est)
```

## Ejercicio 5

Considere un sistema en el cual:

 * Llegan trabajos a tasa $\lambda$ cada un tiempo exponencial.
 * Los trabajos se sirven "todos juntos", una vez que se juntan $K$ trabajos. Esto lleva un tiempo exponencial de  media $1/\mu$. Durante este tiempo no se aceptan nuevos trabajos.
 
1. Analice $X(t) = $cantidad de trabajos en el sistema como una cadena de Markov de tiempo continuo. Hallar la matriz $Q$ y la distribución de régimen. 

2. Simular la cadena anterior para $K=10$.

3. Discutir cuál sería el "retardo medio" que ve un trabajo en este sistema.

### Parte 1

Se plantea la cadena de la siguiente manera:
![Ejemplo](./Imagenes/Ejercicio5_1.jpg)

De las ecuaciones de balance se desliga:
$$
\lambda\pi_0 = \mu\pi_k \\
\pi_0 = \pi_1 \\
\pi_1 = \pi_2 \\
\vdots \\
\lambda\pi_{k-1} = \mu\pi_k
$$
Tambien:
$$
\pi_0+\pi_1+\pi_2+\dots+\pi_k = 1
$$
Entonces:
$$
k\pi_0 +\pi_k = 1 \quad y \quad \pi_k = \frac{\lambda}{\mu}\pi_0
$$
Lo que resulta en:
$$\pi_i = \left\{\begin{array}{cc} 
\frac{1}{k+\frac{\lambda}{\mu}}\quad i\in [0,k-1] \\
\frac{\frac{\lambda}{\mu}}{k+\frac{\lambda}{\mu}}\qquad i=k
\end{array}\right.
$$

### Parte 2

(Se reutiliza la funcion ON-OFF vista con algunas modificaciones)

```julia
function funcEntregas(lambda, mu, pi0, Tfinal, K)
    #lambda: tasa de OFF->ON
    #mu: tasa de ON->OFF
    #pi: distribución para sortear el estado inicial
    #Tfinal: tiempo de simulación
    
    d0 = Categorical(pi0)
    
    d1 = Exponential(1/lambda)
    d2 = Exponential(1/mu)
    
    x=rand(d0)-1
    t=0.0
    aux = x
    
    X=[x];
    T=[t];
    
    while t<Tfinal
        if aux < K
            dt = rand(d1)
            x=x+1;
            aux += 1
        else
            dt = rand(d2)
            x = 0
            aux = 0
        end
        t=t+dt
        push!(T,t)
        push!(X,x)
    end
    
    return T,X
    
end

#Utilización de la función definida
#func_ON_OFF(lambda, mu, pi0, Tfinal)
T,X = funcEntregas(1.0,1.0,ones(11)/11,50,10)

plot(T,X, lt=:steppost, legend=:none)
#scatter!(T,X)
```

### Parte 3

El "retardo medio" que ve un trabajo en este sistema sería:

$$ \bar{W} = \sum_{i=0}^{K} \pi_i\left(\frac{(K-i)}{\lambda} +\frac{1}{\mu}\right)\\
  = \left(\frac{K}{\lambda}+\frac{1}{\mu}\right)\sum_{i=0}^{K} \pi_i- \frac{1}{\lambda}\sum_{i=0}^{K} i\pi_i\\
  \left(\frac{K}{\lambda}+\frac{1}{\mu}\right)\frac{K-1+\rho}{K+\rho}- \frac{1}{\lambda}\left(\frac{K(K-1)}{2(K+\rho)}+\frac{K\rho}{K+\rho}\right)$$
