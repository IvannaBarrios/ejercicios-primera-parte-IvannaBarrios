## Pregunta 1 (multiplexado de capacidad).

Supongamos que tenemos un sistema al que llegan clientes con tasa $N\lambda$. Cada cliente tiene un trabajo de tiempo exponencial de parámetro $\mu$. Compare las siguientes situaciones:

 * Los clientes al llegar son enviados a una de $N$ colas al azar, donde cada una es atendida por un servidor de velocidad $1$.
 
 * Los clientes son todos puestos en una única cola de velocidad $N$.
 
¿Cuál sistema le parece mejor? ¿Qué puede decir sobre el retardo medio en uno y otro sistema?

**Nota:** la respuesta a esta pregunta es la base de las redes "de datagramas" como Internet (Kleinrock, 1969).

## Respuesta:

A simple vista parece ser que en ambos sistemas el cliente veria la misma demora.
### En el sistema *1*: 
Este se comporta como $N$ sistemas individuales $M/M/1$ que estan vinculados por la cola de entrada de tasa $N\lambda$.\
Asi es que cada uno de estos sistemas vera una tasa de entrada de $\lambda$.

Entonces como se vio en la calse, si se analiza a los sistemas de manera individual:
$$\pi_k = (1-\rho)\rho^k,\quad donde \quad \rho = \frac{\lambda}{\mu}$$
Tambien:
$$E[X(t)] =\frac{\rho}{1-\rho}$$
$$E[W(t)]= \frac{1}{\mu - \lambda}$$

Si se analiza el sistema completo, entonces, se debe cosiderar que la ocupacion sera $N$ veces mayor:
$$E[X(t)] =N\frac{\rho}{1-\rho}$$
El retardo medio es el de cada sistema individual, por lo que se mantiene igual:
$$E[W(t)]= \frac{1}{\mu - \lambda}$$

### En el sistema *2*:

Este se comporta como un sistema $M/M/1$ con llegadas exponeciales a la cola de parametro $N\lambda$.\
Tambien el servidor tiene la capacidad de procesar los trabajos con una tasa $N\mu$.
Analizando el sistema como se vio en clase:
$$\pi_k = (1-\rho)\rho^k,\quad donde \quad \rho = \frac{N\lambda}{N\mu}=\frac{\lambda}{\mu}$$
Igualmente:
$$E[X(t)] =\frac{\rho}{1-\rho}$$
$$E[W(t)]= \frac{1}{N\mu - N\lambda} = \frac{1}{N(\mu-\lambda)}$$

El *sistema 2* parece ser el mejor. Es más veloz para los clientes y su funcionamiento será mas confiable a lo largo del tiempo. 

## Ejercicio 2

Se considera un sistema de fila de espera en el cual la cantidad de servidores se adapta a la carga. En particular consideremos el siguiente modelo simplificado:

* Los trabajos llegan como un proceso de Poisson de intensidad $\lambda$.
* Todos los trabajos tienen una duración $\exp(\mu)$.
* Si hay menos de $K$ trabajos (en servicio + en espera), se mantiene un único servidor encendido.
* Si hay $K$ trabajos o más, se enciende un segundo servidor idéntico.

1. Realice un modelo de Markov para este proceso. Analice su estabilidad
2. Para el caso $\lambda=\mu=1$, calcule la distribución invariante del sistema.
3. Deduzca (analítica o numéricamente) la ocupación media del sistema y el retardo medio.
4. Investigue cuál es el *máximo* valor de $K$ que cumple que $\bar{W}<4$.
5. Simule el sistema y estudie su comportamiento empírico para validar los cálculos anteriores.

### Parte 1

El diagrama del modelo de Markov para este proceso, sería el siguiente:
![Ejemplo](./Imagenes/Ejercicio2_2.jpg)
De donde se obtienen las siguientes ecuaciones en equilibrio:
$$\begin{align} \pi_0\lambda &= \pi_1\mu \\ \pi_1\lambda &= \pi_2\mu  \\ \pi_2\lambda &= \pi_3\mu \\ \cdots& \\ \pi_{K-1}\lambda  &= \pi_K2\mu  \\ \cdots& \\ \pi_{i}\lambda  &= \pi_{i+1}2\mu\end{align}$$\
Se observa que similarmente al caso $M/M/m$ para lograr estabilidad la condicion sera $\lambda < 2\mu$

### Parte 2

Las ecuaciones ahora seran:\
$$\begin{align} \pi_0 &= \pi_1\\ \pi_1 &= \pi_2  \\ \pi_2 &= \pi_3 \\ \cdots& \\ \pi_{K-1} &= 2\pi_K \\ \pi_{K} &= 2\pi_{K+1} \\ \cdots& \\ \pi_{i}  &= 2\pi_{i+1}\end{align}$$

De donde se concluye:
$$\pi_i: \left\{\begin{array}{cc}\pi_i = \pi_0 &, i<K\\ \pi_i=\pi_0\left(\frac{1}{2}\right)^{i-(K-1)} &,i \geq K\end{array}\right.$$

Imponiendo $\sum_i \pi_i =1$ entonces se tiene:
$$\sum_{i=0}^{\infty}\pi_i = K\pi_0 + \sum_{i=k}^{\infty} \pi_0\left(\frac{1}{2}\right)^{i-(K-1)} = 1$$
Agrupando y realizando un cambio de variable:
$$\pi_0\left[K + \sum_{j=1}^{\infty}\left(\frac{1}{2}\right)^{j}\right] = \pi_0\left[K + \frac{1}{1-1/2} -1 \right]=1$$
Finalmente,
$$\pi_0(K+1)=1 \quad \Rightarrow \quad \pi_0 = \frac{1}{K+1}$$
El resultado final es:
$$\pi_i: \left\{\begin{array}{cc}\pi_i = \frac{1}{K+1} &, i<K\\ \pi_i=\frac{1}{K+1}\left(\frac{1}{2}\right)^{i-(K-1)} &,i \geq K\end{array}\right.$$

### Parte 3

La ocupacion media en el sistema se calcula:

$$\begin{align}
E[X(t)] &= \sum_{i=0}^\infty i \pi_i = \frac{1}{K+1}\frac{K(K-1)}{2}+ \frac{1}{K+1}\sum_{i=K}^{\infty}i\left(\frac{1}{2}\right)^{i-(K-1)}\\
&= \frac{K(K-1)}{2(K+1)}+ \frac{1}{K+1}\sum_{j=1}^{\infty}(j+K-1)\left(\frac{1}{2}\right)^j\\
&= \frac{K(K-1)}{2(K+1)}+ \frac{K-1}{K+1}+\frac{1}{K+1}\sum_{j=1}^{\infty}j\left(\frac{1}{2}\right)^j\\
&= \frac{K(K-1)}{2(K+1)}+ \frac{K-1}{K+1}+\frac{2}{K+1} = \frac{1}{2}\left[\frac{K(K-1)}{K+1}+ \frac{2K-2}{K+1}+\frac{4}{K+1}\right]\\
&= \frac{K^2-K+2K-2+4}{2(K+1)}=\frac{K^2+K+2}{2(K+1)}\\
\end{align}$$

Y luego segun la ley de Little se puede decir que el retardo medio del sistema será:

$$E[W(t)]=E[X(t)]/\lambda$$

E imponiendo $\lambda = 1$ entonces:

$$E[W(t)]=E[X(t)]$$

### Parte 4 

Se tiene ahora que,
$$\frac{K^2+K+2}{2(K+1)}<4\quad o \quad \frac{K^2+K+2}{(K+1)}<8$$
Planteando la igualdad limite se obtiene el polinomio:
$$K^2-7K-6=0$$
La raiz mayor de este polinomio es $K\approx 7.7$, pero $K$ debe ser un entero, asi que el maximo valor que cumplira la condicion es $K=7$


### Parte 5
Modificando la funcion del caso $M/M/m$ para adaptarla al nuevo se obtiene:

```julia
using Distributions

function funcEj2(lambda,mu,K,x0,Tfinal)
    
    d1 = Exponential(1/lambda)
    d2 = Exponential(1/mu)
    
    x=x0
    t=0.0
    
    T=[t]
    X=[x]
    
    arribos = Int64[]
    partidas = Int64[]
    
    i=1
    
    while t<Tfinal
    
        i=i+1
        if x>0
            t1 = rand(d1)  #Tiempo para subir
            if x<K         #Tiempo para bajar
                t2 = rand(d2)
            else
                t2 = minimum(rand(d2,2))  
            end
            if t1<t2    #Gana subir
                x=x+1
                t=t+t1
                push!(arribos,i)
            else        #Gana bajar
                x=x-1
                t=t+t2
                push!(partidas,i)
            end

        else    
            t=t+rand(d1) #En 0 solo subo
            x=x+1
            push!(arribos,i)
        end
        
        push!(T,t)
        push!(X,x)
        
        
    end

    return T,X,arribos, partidas
end
```

Utilizando la función: 

```julia
using Plots, LaTeXStrings
lambda = 1.0
mu = 1.0
K=7

Tfinal = 10000

T,X,arribos,salidas = funcEj2(lambda,mu,K,0,Tfinal)

plot(T,X,lt=:steppost,xlabel=L"t", ylabel=L"X(t)", label=:none)
```

```julia
Xbar = sum(X[1:end-1].*diff(T))/T[end]
Xbar_teo = (K^2+K+2)/(2*(K+1))
println("Ocupacion media experimento: $Xbar")
print("Ocupacion media teórica: $Xbar_teo")
```

## Ejercicio 3

Un concentrador telefónico fue diseñado para operar con $C = 30$ circuitos. Las llamadas llegan a tasa
$\lambda = 300$ llamadas por hora y tienen una duración media de $3$ min.

 1. Calcule usando la recursión de Erlang provista la probabilidad de bloqueo del sistema. Coteje con la [tabla de Erlang](https://www.pitt.edu/~dtipper/2110/erlang-table.pdf).
 2. Indique cuál es la tasa máxima de llamadas que puede recibir el sistema, asumiendo que la duración se mantiene, sin que el grado de servicio supere el 0.5\% ($B<0.005$).
 3. Debido a un cambio en la tarifa, las llamadas pasan a durar en media $5$ min. Calcule cuántos circuitos se deben agregar para mantener el grado de servicio.
 4. Simule el sistema y verifique lo obtenido anteriormente.

```julia
function erlangB(A,C)
    
    aux = 1       #Calcula recursivamente el termino del numerador
    sum = aux     #Calcula la suma del denominador
    
    for i=1:C
        aux = aux * A/i
        sum=sum+aux
    end
    
    return aux/sum
end
```

### Parte 1

```julia
C = 30 #Capacidad maxima

lambda = 300                  #Frecuencia de ingresos, horas^-1
duracion_llamada = 3          #Duracion, minutos
mu = 1/(duracion_llamada/60)  #Frecuencia de salidas, horas^-1

A = lambda/mu 

B_erlang = erlangB(A,C)

println("Calculando B por recursion da: $(B_erlang*100)%")
print("La tabla muestra aproximadamente que se encuentra entre 0.01% y 0.05%")
```

### Parte 2

De la tabla se obtiene que el valor de $A$ debe ser estrictamente menor a $19.03$.\
Por lo tanto $\lambda$ debera ser menor a $A\mu=19.03\mu=380.6$

```julia
C = 30 #Capacidad maxima
lambda = 380                  #Frecuencia de ingresos, horas^-1
duracion_llamada = 3          #Duracion, minutos
mu = 1/(duracion_llamada/60)  #Frecuencia de salidas, horas^-1
A = lambda/mu 
B_erlang = erlangB(A,C)
println("Calculando B por recursion da: $(B_erlang*100) %")
```

### Parte 3

El nuevo valor de $A$ es $A=300\times\frac{5}{60}=25\quad$ 

Ahora en la tabla se busca entre $0.01\%$y$0.05\%$

Se obtiene $C=44$ porque con esto la probablilidad quedara en el intervalo deseado.

```julia
C = 44 #Capacidad maxima
lambda = 300                  #Frecuencia de ingresos, horas^-1
duracion_llamada = 5          #Duracion, minutos
mu = 1/(duracion_llamada/60)  #Frecuencia de salidas, horas^-1
A = lambda/mu 
B_erlang = erlangB(A,C)
println("Calculando B por recursion da: $(B_erlang*100)%")
```

```julia
using Distributions

function funcMMCC(lambda,mu,C,x0,Tfinal)
    
    d1 = Exponential(1/lambda)
    d2 = Exponential(1/mu)
    
    x=x0
    t=0.0
    
    T=[t]
    X=[x]
    
    arribos = 0
    bloqueos = 0
    
    i=1
    
    while t<Tfinal
    
        i=i+1
        if C>x>0
            t1 = rand(d1)  #Tiempo para subir
            t2 = minimum(rand(d2,x))  #Tiempo para bajar, el minimo de x exponenciales mu.
            
            if t1<t2    #Gana subir
                x=x+1
                t=t+t1
                arribos += 1
            else        #Gana bajar
                x=x-1
                t=t+t2
            end
        elseif x==C
            t1 = rand(d1)  #Tiempo para subir
            t2 = minimum(rand(d2,x))  #Tiempo para bajar, el minimo de x exponenciales mu.
            
            if t1<t2    #Gana subir
                t=t+t1
                bloqueos += 1
            else        #Gana bajar
                x=x-1
                t=t+t2
            end
            
        else    
            t=t+rand(d1) #En 0 solo sube
            x=x+1
            arribos += 1
        end
        
        push!(T,t)
        push!(X,x)
        
        
    end

    return T,X,arribos, bloqueos
end
```

```julia
using Plots, LaTeXStrings
default(size=(1000,500), legendfontsize=12, labelfontsize=12)

C = 30                        #Capacidad maxima
lambda = 300                  #Frecuencia de ingresos, horas^-1
duracion_llamada = 3          #Duracion, minutos
mu = 1/(duracion_llamada/60)  #Frecuencia de salidas, horas^-1

Tfinal = 100

T,X,arribos,bloqueos = funcMMCC(lambda,mu,C,0,Tfinal)

plot(T,X,lt=:steppost,xlabel=L"t", ylabel=L"X(t)", label=:none)
```

```julia
println("B del experimento es: $(100*bloqueos/arribos)%")
```
